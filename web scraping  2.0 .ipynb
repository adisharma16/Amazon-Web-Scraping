```python
from bs4 import BeautifulSoup as bs 
import requests 
import csv
import lxml
```

```python
url =  "https://www.amazon.in/boAt-Airdopes-300-Multipoint-Connection/dp/B0D3DH8TSC/?_encoding=UTF8&pd_rd_w=pcEqI&content-id=amzn1.sym.14ff1b43-540a-4bfd-a18c-350bea29dfa2&pf_rd_p=14ff1b43-540a-4bfd-a18c-350bea29dfa2&pf_rd_r=6JFJCCY738CMFGR8CBKK&pd_rd_wg=UNlY4&pd_rd_r=6666e7a0-0c3b-481b-9d69-2222c1f5e02e&ref_=pd_hp_d_atf_Headphones&th=1"
header = {"user-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/148.0.0.0 Safari/537.36 Edg/148.0.0.0"}
```

```python
reponse = requests.get(url , headers = header)
```

```python
if reponse.status_code == 200:
    html_content = reponse.text
else:
    print("error")
```

```python
## print(html_content)
```

```python
soup = bs(html_content, "html.parser")
```

```python
product_name = soup.find("span" , id = "productTitle").text.strip()
```

```python
print(product_name)
```

```python
product_price = soup.find("span" , class_ = "a-price-whole").text.strip()
```

```python
print(product_price)
```

```python
delivery_location = soup.find("div", id="contextualIngressPtLabel_deliveryShortLine").text.strip()
```

```python
print(delivery_location)
```

```python
Or_fastest_delivery = soup.find("span", class_="a-text-bold").text.strip()
```

```python
print(Or_fastest_delivery)
```

```python
In_stock = soup.find("span", class_="a-size-medium a-color-success primary-availability-message").text.strip()
```

```python
print(In_stock)
```

```python
About_this_item = soup.find("ul",class_="a-unordered-list a-vertical a-spacing-mini").text.strip()
```

```python
print(About_this_item)
```

```python
with open("output.csv" , mode = "w", newline = '', encoding = "utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(['product_name','product_price' , 'In_stock' , 'delivery_location', 'About_this_item'])
    writer.writerow([product_name,product_price, In_stock , delivery_location, About_this_item])
    print('data saved')
```
